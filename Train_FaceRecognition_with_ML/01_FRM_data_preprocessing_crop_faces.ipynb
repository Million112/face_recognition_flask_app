```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
fpath = glob('./data/female/*.jpg')
mpath = glob('./data/male/*.jpg')
```

```python
print('The number of images in Female folder = ', len(fpath))
print('The number of images in Male folder = ', len(mpath))
```

```python
#step 1: read image and convert to RGB
img = cv2.imread(fpath[19]) #read img in BGR
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert BGR to RGB

#step 2: aplly haar cascade classifier
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
faces_list = haar.detectMultiScale(gray, 1.5, 5)
print(faces_list)
for x,y,w,h in faces_list:
    cv2.rectangle(img_rgb, (x,y), (x+w, y+h), (0,255,0), 2)
    #step 3: crop face
    roi = img_rgb[y:y+h, x:x+w]

    plt.imshow(roi)
    plt.axis('off')
    plt.show()


#step 4: save image
```

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
for i in range(len(fpath)):
    try:

        #step 1: read image and convert to RGB
        img = cv2.imread(fpath[i]) #read img in BGR
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert BGR to RGB
        
        #step 2: aplly haar cascade classifier
        gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray, 1.6, 5)
        for x,y,w,h in faces_list:
            #step 3: crop face
            roi = img[y:y+h, x:x+w]
            #step 4: save image
            cv2.imwrite(f'./crop_data/female/female_{i}.jpg', roi)
            print('Image Sucessfully processed')
    except:
        print('Unable to Process the image')
        
```

```python
for i in range(len(mpath)):
    try:

        #step 1: read image and convert to RGB
        img = cv2.imread(mpath[i]) #read img in BGR
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert BGR to RGB
        
        #step 2: aplly haar cascade classifier
        gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray, 1.4, 5)
        for x,y,w,h in faces_list:
            #step 3: crop face
            roi = img[y:y+h, x:x+w]
            #step 4: save image
            cv2.imwrite(f'./crop_data/male/male_{i}.jpg', roi)
            print('Image Sucessfully processed')
    except:
        print('Unable to Process the image')
```
