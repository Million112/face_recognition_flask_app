```python
import cv2
from face_recognition import faceRecognitionPipeline
```

```python
import matplotlib.pyplot as plt
```

```python
pred_img, pred_dict =  faceRecognitionPipeline('./test_images/top_2.jpg')
```

```python
cv2.imshow("Hinh anh", pred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_rgb = cv2.cvtColor(pred_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(img_rgb)
plt.axis('off')
plt.show() 


#gen report
for i in range(len(pred_dict)):
    obj_gray = pred_dict[i]['roi']
    obj_eig = pred_dict[i]['eig_img'].reshape(300,300)
    plt.subplot(1,2,1)
    plt.imshow(obj_gray, cmap = 'gray')
    plt.title('Gray ScaleImage')
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(obj_eig, cmap = 'gray')
    plt.title('Eigen Image')
    plt.axis('off')

    plt.show()
    print('Predicted Gender = ', pred_dict[i]['prediction_name'])
    print('Predicted score = {:,.2f} %'.format(pred_dict[i]['score']*100))

    print('-'*100)
```

```python
cap = cv2.VideoCapture('./test_images/video.mp4')

while True:
    ret, frame = cap.read()

    if ret == False:
        break

    pred_img, pred_dict = faceRecognitionPipeline(frame, path = False)

    cv2.imshow('video', pred_img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
